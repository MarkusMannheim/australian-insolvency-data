```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_json("insolvencyData.json")
data["week"] = data["date"].dt.week
data["count"] = [1] * len(data)
def convert_week(x):
    if x > 13:
        x = x - 52
    return x
data["week"] = data["week"].apply(convert_week)
data["type"].describe()
```

```python
week_count = data.groupby("week").sum()
week_count
```

```python
plt.bar(week_count.index, week_count["count"])
plt.show()
```
